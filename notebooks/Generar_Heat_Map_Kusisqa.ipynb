```python
from datasets.kusisqadim import KusisqaDim

dataset = KusisqaDim(
        "../ser_datasets/kusisqadim/train.csv",
        "../ser_datasets/kusisqadim/audios"
)

x,y = zip(*list(dataset))
```

```python
import torch

labels = torch.stack(list(y))
# print(labels)

valence = labels[:,0]
arousal = labels[:,1]
dominance = labels[:,2]
```

## Arousal vs Valence

```python
import numpy as np
import seaborn as sns
valence_arousal = np.zeros((5,5),dtype=np.int32)

for i in range(len(valence)):
    valence_arousal[int(valence[i])-1][int(arousal[i])-1] += 1
  


ax = sns.heatmap(valence_arousal, annot=True, fmt="d")
ax.invert_yaxis()
ax.set_ylabel('Valence')
ax.set_xlabel('Arousal')
```

```python
ax = sns.kdeplot(arousal,valence, shade = True, cmap="rocket_r")
ax.plot(arousal, valence, 'o', alpha=0.05 ,color='black')
ax.set_xlabel('Valence')
ax.set_ylabel('Arousal')
```

## Arousal vs Dominance

```python
dominance_arousal = np.zeros((5,5),dtype=np.int32)

for i in range(len(valence)):
    dominance_arousal[int(dominance[i])-1][int(arousal[i])-1] += 1

print(dominance_arousal)

ax = sns.heatmap(dominance_arousal, annot=True, fmt="d")
ax.invert_yaxis()
ax.set_ylabel('Dominance')
ax.set_xlabel('Arousal')
```

```python
ax = sns.kdeplot(arousal,dominance, shade = True, cmap="rocket_r")
ax.plot(arousal, dominance, 'o', alpha=0.05 ,color='black')
ax.set_xlabel('Dominance')
ax.set_ylabel('Arousal')
```
